```python
import pandas as pd
import firebase_admin
from firebase_admin import credentials, firestore
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor

# 1. 이메일 읽기
with open('received_email.txt', 'r') as file:
    email = file.read().strip()
    print("이메일:", email)

# 2. Firebase Admin SDK 초기화
if not len(firebase_admin._apps):
    cred = credentials.Certificate('C:/pododoc/key/kosmo-96bbe-60906db745e9.json')
    firebase_admin.initialize_app(cred)

# 3. Firestore 클라이언트 생성
db = firestore.client()

# 4. 이메일로 데이터 조회하는 함수 정의
def fetch_data_by_email(email):
    reviews_ref = db.collection('review')
    query = reviews_ref.where('email', '==', email)
    results = query.stream()
    
    data = []
    for doc in results:
        doc_data = doc.to_dict()
        if 'index' in doc_data and 'rating' in doc_data:
            data.append({
                'index': doc_data['index'],
                'rating': doc_data['rating']
            })
    
    return data

# 5. 이메일에 맞는 데이터 가져오기
data = fetch_data_by_email(email)
print("Firestore에서 가져온 데이터:")
print(data)

# 6. Combined_Wine_Data.csv 파일 읽기
wine_data = pd.read_csv('data/Combined_Wine_Data.csv')

# 7. Red wine만 필터링
red_wine_data = wine_data[wine_data['wine_type'] == 'Red wine']

# 8. index를 기준으로 body, texture, sweetness, acidity, flavor1, flavor2, flavor3 컬럼 데이터 추출
def get_wine_attributes(index_list):
    attributes = []
    for idx in index_list:
        row = red_wine_data[red_wine_data['index'] == idx]
        if not row.empty:
            attributes.append({
                'index': idx,
                'body': row['body'].values[0],
                'texture': row['texture'].values[0],
                'sweetness': row['sweetness'].values[0],
                'acidity': row['acidity'].values[0],
                'flavor1': row['flavor1'].values[0],
                'flavor2': row['flavor2'].values[0],
                'flavor3': row['flavor3'].values[0]
            })
    return attributes

# Firestore 데이터에서 추출한 index 번호 리스트
index_list = [entry['index'] for entry in data]

# 와인 데이터에서 속성 추출
attributes = get_wine_attributes(index_list)

# 9. Firestore 데이터와 와인 데이터 결합
final_data = []
for entry in data:
    index = entry['index']
    rating = entry['rating']
    # attributes 리스트에서 해당 index에 맞는 데이터 찾기
    attribute_data = next((attr for attr in attributes if attr['index'] == index), None)
    if attribute_data:
        final_data.append({
            'index': index,
            'rating': rating,
            'body': attribute_data['body'],
            'texture': attribute_data['texture'],
            'sweetness': attribute_data['sweetness'],
            'acidity': attribute_data['acidity'],
            'flavor1': attribute_data['flavor1'],
            'flavor2': attribute_data['flavor2'],
            'flavor3': attribute_data['flavor3']
        })

# 최종 데이터를 DataFrame으로 변환
df_final = pd.DataFrame(final_data)

# 10. 별점 기반 가중치 컬럼 추가
df_final['weight'] = df_final['rating']  # 별점 자체를 가중치로 사용

# 11. 범주형 데이터 전처리
label_encoders = {}
for col in ['flavor1', 'flavor2', 'flavor3']:
    le = LabelEncoder()
    df_final[col] = le.fit_transform(df_final[col])
    label_encoders[col] = le

# 12. 데이터와 타겟 분리
features = df_final.drop(['index', 'rating', 'weight'], axis=1)
target = df_final['rating']
weights = df_final['weight']  # 가중치로 사용될 컬럼

# 13. 정규화
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# 14. 데이터 분할
X_train, X_test, y_train, y_test, w_train, w_test = train_test_split(features_scaled, target, weights, test_size=0.2, random_state=42)

# 15. 모델 정의 및 학습
model = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, random_state=42)
model.fit(X_train, y_train, sample_weight=w_train)  # sample_weight를 사용하여 가중치 적용

# 16. 모델 평가
score = model.score(X_test, y_test)
print(f"모델 성능 (R^2 score): {score}")
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor

# 예제 데이터 생성 및 모델 학습을 위한 코드
# ... (이전 코드와 동일, 학습 데이터 및 모델 정의)

# 1. 범주형 변수를 인코딩하는 함수 정의
def preprocess_input_data(body, texture, sweetness, acidity, flavor1, flavor2, flavor3):
    # 입력 데이터 전처리
    # Check if the flavor is in the trained encoder
    if flavor1 not in label_encoders['flavor1'].classes_:
        flavor1_encoded = -1  # Or any default value indicating "unknown"
    else:
        flavor1_encoded = label_encoders['flavor1'].transform([flavor1])[0]
    
    if flavor2 not in label_encoders['flavor2'].classes_:
        flavor2_encoded = -1
    else:
        flavor2_encoded = label_encoders['flavor2'].transform([flavor2])[0]
    
    if flavor3 not in label_encoders['flavor3'].classes_:
        flavor3_encoded = -1
    else:
        flavor3_encoded = label_encoders['flavor3'].transform([flavor3])[0]
    
    # 전처리된 데이터 반환
    return pd.DataFrame({
        'body': [body],
        'texture': [texture],
        'sweetness': [sweetness],
        'acidity': [acidity],
        'flavor1': [flavor1_encoded],
        'flavor2': [flavor2_encoded],
        'flavor3': [flavor3_encoded]
    })

# 예측 함수 정의
def predict_wine_score(body, texture, sweetness, acidity, flavor1, flavor2, flavor3):
    # 입력 데이터 전처리
    input_data = preprocess_input_data(body, texture, sweetness, acidity, flavor1, flavor2, flavor3)
    
    # 데이터 정규화
    input_data_scaled = scaler.transform(input_data)
    
    # 예측
    predicted_score = model.predict(input_data_scaled)
    
    return predicted_score[0]

# 임의의 데이터로 예측
example_score = predict_wine_score(85, 40, 50, 1, 'oaky', 'red fruit', 'citrus')
print(f'예상점수: {example_score}')
```
