```python
max = df['wine_rating'].max()
max
min = df['wine_rating'].min()
min
```

```python
x = list(df['index'])
x
y = list(df['wine_rating'])
y
colors=['r', 'g', 'b', 'k', 'c', 'm', 'y', 'orange']
bar = plt.bar(x, y, color = colors, alpha=0.5, width=0.9)
plt.ylim(min, 5)
plt.xticks(rotation=105) #x 축의 이름 각도를 45도로 설정
plt.yticks(rotation=105) #y 축의 이름 각도를 45도로 설정
bar[0].set_hatch('/')
bar[1].set_hatch('x')
bar[2].set_hatch('..')
bar[3].set_hatch('\\')
bar[4].set_hatch('//')
bar[5].set_hatch('\\')
bar[6].set_hatch('/')
bar[7].set_hatch('.')
plt.show()
```
