```python
import numpy as np
import pandas as pd

# 랜덤 시드 설정 (재현 가능성 확보)
np.random.seed(0)

# 데이터 생성
num_samples = 50
body = np.random.uniform(1, 10, num_samples)   # Body 값: 1부터 10까지
texture = np.random.uniform(1, 10, num_samples) # Texture 값: 1부터 10까지
sweetness = np.random.uniform(1, 10, num_samples) # Sweetness 값: 1부터 10까지
score = np.random.uniform(0, 5, num_samples)   # Score 값: 0부터 5까지

# DataFrame 생성
df = pd.DataFrame({
    'body': np.round(body, 1),
    'texture': np.round(texture, 1),
    'sweetness': np.round(sweetness, 1),
    'score': np.round(score, 1)
})

# CSV 파일로 저장
df.to_csv('random_data1.csv', index=False)

print("데이터가 'random_data1.csv'에 저장되었습니다.")
print(df.head())  # 데이터 확인
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np
import matplotlib.pyplot as plt

# 데이터 로드
df = pd.read_csv('random_data.csv')

# 입력 변수와 목표 변수 설정
X = df[['body', 'texture', 'sweetness']]
y = df['score']

# 데이터 분할 (80% 학습, 20% 테스트)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# 모델 생성
model = LinearRegression()

# 모델 학습
model.fit(X_train, y_train)

# 예측 수행
y_pred = model.predict(X_test)

# 모델 평가 (평균 제곱 오차 계산)
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse:.2f}")
```

```python
# 새로운 데이터 예측
new_data = pd.DataFrame({
    'body': [3.5, 7.2],
    'texture': [4.0, 2.9],
    'sweetness': [5.0, 3.5]
})

# 예측 수행
predictions = model.predict(new_data)
print(f"New Data Predictions: {predictions}")

# 실제 vs 예측 시각화
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.7, edgecolors='w', s=100)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.title('Actual vs. Predicted Scores')
plt.xlabel('Actual Score')
plt.ylabel('Predicted Score')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np
import matplotlib.pyplot as plt

# 데이터 로드
df = pd.read_csv('random_data.csv')

# 입력 변수와 목표 변수 설정
X = df[['body', 'texture', 'sweetness']]
y = df['score']

# 데이터 분할 (80% 학습, 20% 테스트)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# 모델 생성
model = LinearRegression()

# 모델 학습
model.fit(X_train, y_train)

# 예측 수행
y_pred = model.predict(X_test)

# 모델 평가 (평균 제곱 오차 계산)
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse:.2f}")

# 새로운 데이터 예측
# 각각의 변수에 대해 개별적으로 예측
new_data = pd.DataFrame({
    'body': [3.5],
    'texture': [4.0],
    'sweetness': [5.0]
})

# 예측 수행
prediction = model.predict(new_data)
print(f"Prediction for body=3.5, texture=4.0, sweetness=5.0: {prediction[0]:.2f}")

# 시각화
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.7, edgecolors='w', s=100)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.title('Actual vs. Predicted Scores')
plt.xlabel('Actual Score')
plt.ylabel('Predicted Score')
plt.grid(True)
plt.show()
```
